```python
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create instances of Gaussian Naive Bayes and Random Forest classifiers
gnb = GaussianNB()
rf = RandomForestClassifier()

# Create a voting classifier with GaussianNB and Random Forest as estimators
voting_classifier = VotingClassifier(estimators=[('gnb', gnb), ('rf', rf)], voting='hard')

# Train the voting classifier on the training data
voting_classifier.fit(symptoms_train.values, prognosis_train.values.ravel())
# Make predictions using the voting classifier
predictions = voting_classifier.predict(symptoms_test.values)

# Evaluate the performance of the ensemble model
accuracy = accuracy_score(prognosis_test.values.ravel(), predictions)  
print("Accuracy:", accuracy)
```
